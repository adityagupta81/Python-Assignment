# Task 6 --> Convert the budget and gross values from ‚Äúdollars‚Äù to ‚Äúdollars in millions‚Äù for all top-rated movies. Round the converted values down to 3 decimal places. For example, a value of 192,345,273 should be converted to 192.345.

```python
# Create and filter the budget cols and round to 3 decimal place 
df["budget"] =np.round (df ['budget'] / 1000000,2)
df['budget']
```

# Task 7 -->List all details for the top 10 movies with the highest profit, sorted from highest to lowest. Hint: Profit is not a column in the DataFrame. You will need to calculate it.

```python
# Calculating profit by subtracting budget from gross cols.
df['Profit'] = df['Gross'] - df['budget']

# Display Profit cols.
df.head(10)
```

```python
# Sorting Values by profit and arrange them from highest to lowest.
df= df.sort_values (by='Profit', ascending = False)

# Display the details of top 10 movies with highest profit.
df.head(10)
```

# Task 8 --> Generate a list of all the actors, in alphabetical order by the first name, that have starred in a top-rated movie. If an actor has starred in multiple movies, their name should appear only once on the list. Assume that all actors‚Äô names are in the format <first_name> <last_name>.

```python
# Collect  all the values of the cols of  actor_1_name, actor_2_name and  actor_3_name in a single series and called them as all_actors and sort them .

actor_1 = df['actor_1_name']
actor_2 = df['actor_2_name']
actor_3 = df['actor_3_name']

all_actors = actor_1.append(actor_2).append(actor_3).sort_values()

# Display  the name of all actors. 
all_actors
```

```python
# Convert the series into a list after removing the duplicates from all actors name using unique function.
all_actor_list = list(all_actors.unique())
all_actor_list
```

```python
# Removing 0from the above list by using remove function.
all_actor_list.remove('0')
all_actor_list
```

# Task 10 -->Create a data visualization that shows each country and the number of top-rated movies produced in it. Find the country that produced the most top-rated movies. 

```python
# Creating a data visualization by group and sorting them by Title.
Data_visua_df = df.groupby(['Country']).count()[['Title']].sort_values(by='Title', ascending = False)

# Display the Country and Title.
Data_visua_df
```

```python
# Creating a bar plot for the above data.
Data_visua_df.plot(kind ='bar')
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()

# Find the Country with the most top Rated movies .
Top_rated_country = Data_visua_df.idxmax()
print(f"Country with the most Top Rated movies is {Top_rated_country[0]} with {Data_visua_df['Title'].max()} movies.")
```
